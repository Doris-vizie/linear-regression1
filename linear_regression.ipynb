```python
# linear regression using ordinary least mean square method.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('dataset.csv')
dataset.head()
```

```python
print(dataset.shape)
```

```python
#initialize our inputs and outputs
X = dataset['Head Size(cm^3)'].values
Y = dataset['Brain Weight(grams)'].values

#mean of our inputs and outputs
x_mean = np.mean(X)
y_mean = np.mean(Y)

#total number of values
n = len(X)

#using the formula to calculate b1 and b0
numerator = 0
denominator = 0

for i in range(n):
    numerator += (X[i] - x_mean) * (Y[i] - y_mean)
    denominator += (X[i] - x_mean)**2

b1 = numerator/denominator
b0 = y_mean - (b1 * x_mean)

#printing the coeffcients
print(b1, b0)
```

```python
#plotting values
x_max = np.max(X) + 100
x_min = np.min(X) - 100

#calculating line values of x and y
x = np.linspace(x_min, x_max, 1000)
y = b0 + b1 * x

#plotting line
plt.plot(x, y, color='#00ff00', label='linear regression')

#plot the data line
plt.scatter(X, Y, color='#ff0000', label='data point')

#x-axis label
plt.xlabel('Head Size(cm^3)')

#y-axs label
plt.ylabel('Brain Weight(grams)')

plt.legend()
plt.show()
```

```python
#checking for accuracy using root mean square error
rmse = 0
for i in range(n):
    y_pred = b0 + b1* X[i]
    rmse += (Y[i] - y_pred)**2
rmse = np.sqrt(rmse/n)   
print(rmse)
```
